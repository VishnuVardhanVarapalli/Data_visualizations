```
#neccesary libraries for data analysis and data visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import geopandas as gpd
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```
dataset = pd.read_csv('/content/drive/My Drive/datasets/CoVid -19/CoVid-19 data.gov.in')
```

```
dataset
```

```
fp = "/content/drive/My Drive/datasets/covid19-in-india/Igismap/Indian_States.shp"
map_df = gpd.read_file(fp)
```

```
merged = map_df.set_index('st_nm').join(dataset.set_index('Name of State / UT'))
```

```
merged
```

```
plt.figure(figsize=(30,30))
plt.bar(dataset['Name of State / UT'],dataset['Total Confirmed cases'])
plt.xticks(rotation = 90)
plt.title('Visualization of effect of COVID-19 cases in Different States',color = 'Blue')
plt.xlabel('States got effected by COVID-19 in India',color = 'Red')
plt.ylabel('Total Confirmed cases of COVID-19 State-wise',color = 'Red')
fig = plt.gcf()
fig.set_size_inches(20,6) 
```

As we can see from bar graph above, the most affected states are Delhi,Gujarat,MahaRasthra and Tamil Nadu.As we know that, those 4 states are metropolition cities, highly populated, corona spread will be very drastically high there. so as our finaicial economy is getting down, it's beter to maintain lock down in thoses four states and in remaining places, in containment zones if we mantain lockdown, with that our economy will get back to the normal state.





```
fig, ax = plt.subplots(1, figsize=(10, 6))
ax.axis('off')
ax.set_title('State Wise affect of Corona Cases in India', fontdict={'fontsize': '25', 'fontweight' : '3'})
merged.plot(column='Total Confirmed cases', cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
```

As you can see from the map above. the most affected state is Maharastra.So it's better to maintain a strict lockdown as there. As other states are got lockdown in containment zones ony,it's better to help the maharasthra in financial, in case of food for poor people and things like that.So, with that unity only,we can decrease the affect of this covid-19.

```
plt.figure(figsize=(30,9))
plt.bar(dataset['Name of State / UT'],dataset['Cured/Discharged/Migrated'])
plt.xticks(rotation = 90)
plt.title('Visualization of effect of COVID-19 cases in Different States',color = 'Blue')
plt.xlabel('States got effected by COVID-19 in India',color = 'Red')
plt.ylabel('Cured/Discharged/Migrated cases of COVID-19 State-wise',color = 'Red')
plt.rcParams.update({'font.size': 40})
```

when you see the graph above, the Cured cases are looking somewhat cool in every state other this 4 states i.e., Delhi,Gujarat,Maharasthra and Tamil Nadu. Because, the statics of confirmed and cured cases are very far away from each other. As, other states are showing high progress in curing the Covid, it's better those states should this help this states to get out of this crucial period.

```
fig, ax = plt.subplots(1, figsize=(10, 6))
ax.axis('off')
ax.set_title('State Wise Cured/Discharged/Migrated in India', fontdict={'fontsize': '25', 'fontweight' : '3'})
merged.plot(column='Cured/Discharged/Migrated', cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
```

As we can see from map above, many states are trying to get out of this affect of covid-19, but some of the states like as mentioned above, those 4 states are severly got affected by that. As affect is less in other states, pople of those states should not got roam on roads without any precautions and things. because, pople should remember that the increase in Covid cases is directly proportional to the increase in Deaths.  

```
plt.figure(figsize=(30,30))
plt.bar(dataset['Name of State / UT'],dataset['Death'])
plt.xticks(rotation = 90)
plt.title('Visualization of effect of COVID-19 cases in Different States',color = 'Blue')
plt.xlabel('States got effected by COVID-19 in India',color = 'Red')
plt.ylabel('Deaths of COVID-19 State-wise',color = 'Red')
fig = plt.gcf()
fig.set_size_inches(20,6)
```

As you can see from graph above, the death are very less in almost all states except in Gujarat and maharasthra. Also, in West bengal the dath rate is somewhat high. So, it's better to maintain a complete lockdown there, for every period of time(like 1 Week), we need to check the confirmed cases rate, cured rate and death rate of Every place in West bengal,according to that the lockdown should be removed in less affected places. As we can see that other states are getting looking somewhat less in death rate, we can remove lockdown there.

```
fig, ax = plt.subplots(1, figsize=(10, 6))
ax.axis('off')
ax.set_title('State Wise stats of deaths in India', fontdict={'fontsize': '25', 'fontweight' : '3'})
merged.plot(column='Death', cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
```

```
states=dataset['Name of State / UT']
categories = ['Total Confirmed cases','Cured/Discharged/Migrated', 'Death']
pos = np.arange(len(states))
Total_confirmed_cases = dataset['Total Confirmed cases']
Cured_Discharged_Migrated_cases = dataset['Cured/Discharged/Migrated']
Deaths = dataset['Death']
plt.bar(pos,Total_confirmed_cases,color='Red')
plt.bar(pos,Cured_Discharged_Migrated_cases,color='Green')
plt.bar(pos,Deaths,color='black')
plt.xticks(pos, states)
plt.xlabel('States of India', fontsize=16)
plt.ylabel('Cured/Discharged/Migrated,Deaths', fontsize=16)
plt.title('Total Confirmed cases,Cured/Discharged/Migrated,Deaths in India',fontsize=18)
plt.legend(categories,loc=2,fontsize = 10)
plt.xticks(rotation = 90,fontsize = 10)
plt.yticks(fontsize = 10)
fig = plt.gcf()
fig.set_size_inches(20,6) 
plt.show()
```

As you can see the graph above, almost everstate have confirmed cases greater than the cured cases. Death rate is some what less, but we should not rely on that, to remvoe lockdown,because as I said as cases increases, definitely death rate will increases, so everyone should take precautions which are said by our governments and eveyone should follow the rules very strictly.
